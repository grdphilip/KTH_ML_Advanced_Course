```python

def calculate_similarity_per_row(frame,mep_votes):
    sim_list = frame - mep_votes
    count = np.count_nonzero(sim_list, axis=1)
    normalized = count/len(mep_votes)
    return normalized

def sim_function_1(frame):
    
    votes = frame.iloc[:, 10:]
    votes_array = votes.to_numpy()
    
    n_MEPs = votes_array.shape[0]
    similarity_matrix = np.zeros((n_MEPs,n_MEPs))

    for i in range(n_MEPs):
        similarity_matrix[i] = calculate_similarity_per_row(votes_array,votes_array[i])
        
    similarity_frame = pd.DataFrame(similarity_matrix, index=frame['FullName'], columns=frame['FullName'])
    return similarity_frame
```

```python
distance_matrix1 = sim_function_1(frame)
```

```python
import pandas as pd
from scipy.spatial.distance import pdist, squareform

def hamming_similarity(frame):
    # Select the columns containing the votes
    votes = frame.iloc[:, 10:]
    
    # Calculate the pairwise Hamming distances and convert to similarity scores
    hamming_distances = pdist(votes, metric='hamming')
    hamming_similarity_scores = 1 - squareform(hamming_distances)
    
    # Create a DataFrame for the similarity matrix
    similarity_matrix = pd.DataFrame(hamming_similarity_scores, index=frame['FullName'], columns=frame['FullName'])
    
    return similarity_matrix
```

```python
sim_matrix2 = hamming_similarity(frame)
distance_matrix2 = round(abs(sim_matrix2 - 1),6)
```

```python
double_centered_m1 = distance_matrix1 - distance_matrix1.mean(axis=0)
```

```python
double_centered_m2 = distance_matrix2 - distance_matrix2.mean(axis=0)
```

```python
def get_positions(distance_matrix):
    a = [distance_matrix.mean(axis=1)]
    b = [distance_matrix.mean(axis=0)]
    double_centered_mtrx = distance_matrix - np.transpose(a) - np.transpose(b) + distance_matrix.mean()
    eigenvalues, eigenvectors = np.linalg.eigh(double_centered_mtrx)
    
    sorted_indices = np.argsort(eigenvalues)[::-1]
    top_two_eigenvectors = eigenvectors[:, sorted_indices[:2]]

    coordinates = np.sqrt(np.abs(eigenvalues[sorted_indices[:2]])) * top_two_eigenvectors
    mep_coordinates = coordinates[:, :2]
    mep_positions = pd.DataFrame(mep_coordinates, columns=['X', 'Y'])
    positions = np.transpose(mep_positions.to_numpy())
    
    return positions, mep_positions
```

```python
positions1 , mep_positions1 = get_positions(distance_matrix1)
positions2 , mep_positions2 = get_positions(distance_matrix2)

plot1 = plt.scatter(positions1[0],positions1[1])
plt.title("egen funktion")
```
